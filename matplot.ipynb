3d surface plots

```python
x=np.linspace(-10 ,10 , 100)
y=np.linspace(-10 ,10 ,100)
```

```python
xx ,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z ,cmap='viridis')
```

```python
# for contour plot
ax=plt.subplot()
ax.contour(xx,yy,z,cmap='viridis')
#you can use contourf
```

HEATMAP


```python
delivery=pd.read_csv('/content/deliveries.csv')
delivery.head()
```

```python
#delivery['ball'].unique() getting 9 balls
temp=delivery[delivery['ball'].isin((1,2,3,4,5,6)) & (delivery['batsman_runs']==6)]
```

```python
grid=temp.pivot_table(index='over', columns='ball' , values='batsman_runs' ,aggfunc='count' )
```

```python
plt.figure(figsize=(10,10))
plt.imshow(grid)
plt.colorbar()
```
